```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
```

```python
print(df.shape)  #rows*columns
```

```python
print(df.info())  #info about class, range, no of entries, column names, no of missing values, data-type    
```

```python
print(df.describe())       #statistical summary of numerical columns
```

```python
print(df.isnull().sum())   #count of missing values in each column
```

```python
sns.countplot(x="survived", data=df) # A plot of one of the attributes 'survived' in the dataset
plt.show()
```

```python
sns.countplot(x="sex", hue="survived",data=df)
plt.show()
#Plotting an attribute 'sex' with respect to another attribute 'survived'
```

```python
sns.histplot(df["age"].dropna(), bins=30, kde=True)
plt.show()
#Getting a histogram plot of an attribute 'age' while cleaning up noise (Missing values) and dividing dataset into concise parts
#Plotting wavy line to observe trend
```
